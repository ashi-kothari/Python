```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr())
```

From the above heat map we can now see which variables are poorly correlated and which ones are strongly correlated. Some insights that can be drawn are- (1) Percentage in 10 class are corelated to percentage in 12 class which shows that the performance of students in board exams is quite similar in both the years. (2) Performance in 10 class is corelated to percentage in undergraduation and also percentage_MBA, showing performance of students is constant in all the years. (3) post score is also corelated to pre score, showing students scoring higher marks in pre education are scoring higher marks in post education also.

## Plotting the features against the 'Target' variable


```python
sns.catplot(data=data,kind='count',x='post_score',col='STATE')
```

Students from South Zone have scored highest in post_score as compared to all other States.

```python
sns.catplot(data=data,kind='bar',x='post_score',y='Gender')
```

More number of Males have scored higher marks as compared to females.

```python
plt.figure(figsize=(20,10))
sns.countplot(data['Specialization'],hue=data['post_score'])
plt.show()
```

```python
plt.figure(figsize=[10,20])

plt.subplot(411)
sns.countplot(x=data['Age_in_years'],hue=data['post_score'])
plt.subplot(412)
sns.countplot(x=data['Marital_status'],hue=data['post_score'])
plt.subplot(413)
sns.countplot(x=data['Place_you_belong_to'],hue=data['post_score'])
plt.subplot(414)
sns.countplot(x=data['perceived#Job#Skill'],hue=data['post_score'])
plt.show()
```

This concludes that students of age group 21 and 24 years have scored higher scores as compared to others. Also, there were only few married students also they have scored very less scores as compared to singles. Students from Urban area have scored higher scores as compared to students from any other areas. Also, students from preferred skills have scored higher than any other skill set.

```python
sns.countplot(data['Gender'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Gender')
```

The graph shows that count of males is greater than females in education sector.

```python
fig=plt.figure(figsize=(8,10))
sns.countplot(x='Previous_Degree',hue='post_score',data=data ,color='green')
plt.show()
```

The graph shows that students from science background have scored higher scores than any other graduation background.

```python
sns.catplot(data=data,kind='point',x='pre_score',y='post_score')
```

This shows that higher number of students have scored higher post_score as compared to students scoring high pre_score, also their scores have increased comparatively.
